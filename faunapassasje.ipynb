```python
ant = len( mydf )
col = ['Formål', 'Type']
fylker = list( mydf['fylke'].unique() )
for fylke in fylker: 
    temp = mydf[ mydf['fylke'] == fylke]
    ant = len( temp)
    print( f"=====================\n\n   Fylke {fylke}")
    for cc in col: 
        mangler = temp[cc].isna().sum()
        print( f"\tFylke {fylke}:  {mangler:4} av {ant:4} {round(100*mangler/ant):5}% mangler dataverdi for {cc}" ) 
```
